## Connecting to the Database
Import the necessary packages and connect to the database **data.sqlite**.

```python
#Your code here
import sqlite3
import pandas as pd
connection = sqlite3.connect('data.sqlite', detect_types=sqlite3.PARSE_COLNAMES)
cursor = connection.cursor()
```

## Display the names of all the employees in Boston.

```python
#Your code here
cursor.execute('''SELECT lastName, firstName from employees join offices using(officeCode) where city = 'Boston';''').fetchall()
```

## Do any offices have no employees?

```python
#Your code here
cursor.execute('''select city, count(*) from offices left join employees using(officeCode) group by 1''')
df = pd.DataFrame(cursor.fetchall())
df.head()

# don't understand this?????
```
